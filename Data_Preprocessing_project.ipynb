### Step 1: Import Libraries and Load Data

```python
import pandas as pd
import numpy as np

# Sample messy dataset
data = {
    "CustomerID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "CustomerName": ["Raj", None, "Simran", "Amit", "Priya", "Karan", "Neha", None, "Vikas", "Asha"],
    "Age": [25, None, 30, "unknown", 45, 29, None, 33, "N/A", 40],
    "Price": [500, 1500, None, 2000, 1200, "NA", 800, 1000, None, 2200],
    "Quantity": [2, None, 3, "two", 5, 1, None, "3", 4, 2],
    "Discount": [5, 10, None, "N/A", 15, 0, 5, None, 10, "none"],
    "ReceivedDate": ["2025-08-01", "2025/08/03", None, "03-08-2025", "2025-08-05", "Invalid", "2025-08-07", None, "2025-08-09", "2025-08-10"],
    "DeliveryDate": ["2025-08-05", None, "2025-08-06", "06-08-2025", "2025/08/09", None, "2025-08-12", "Invalid", "2025-08-13", "2025-08-14"],
    "Email": ["raj@example.com", None, "simran@example", "amit@", "priya@example.com", "karan#mail.com", None, "neha@example.com", "vikas@", "asha@example.com"],
    "Review": ["Good", "Bad", None, "Average", "Excellent", "Poor", "Bad", None, "Good", "Excellent"]
}

# Load into DataFrame
df = pd.DataFrame(data)

# Display initial dataset
df
```

### Step 2: Check Missing Values

```python
# Check missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Summary statistics for numeric columns
df.describe()
```

### Step 3: Handle Missing & Invalid Values

```python
# Age column
df["Age"] = df["Age"].replace(["unknown", "N/A"], np.nan)
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Price column
df["Price"] = df["Price"].replace(["NA", "N/A"], np.nan)
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Price"] = df["Price"].fillna(df["Price"].mean())

# Quantity column
df["Quantity"] = df["Quantity"].replace(["two"], 2)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())

# Discount column
df["Discount"] = df["Discount"].replace(["N/A", "none"], np.nan)
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")
df["Discount"] = df["Discount"].fillna(0)
print(df)
```

### Step 4: Clean & Standardize Dates

```python
# Convert date columns to datetime, invalid parsing becomes NaT
df["ReceivedDate"] = pd.to_datetime(df["ReceivedDate"], errors="coerce")
df["DeliveryDate"] = pd.to_datetime(df["DeliveryDate"], errors="coerce")

# Fill missing/invalid dates with a reference date
df["ReceivedDate"] = df["ReceivedDate"].fillna(pd.Timestamp("2025-08-01"))
df["DeliveryDate"] = df["DeliveryDate"].fillna(pd.Timestamp("2025-08-05"))

# Display dataset after cleaning dates
df[["ReceivedDate", "DeliveryDate"]]
```

### Step 5: Clean & Standardize Text Columns

```python
# CustomerName: fill missing, capitalize
df["CustomerName"] = df["CustomerName"].fillna("Unknown")
df["CustomerName"] = df["CustomerName"].str.title().str.strip()

# Email: fill missing, lowercase, remove spaces
df["Email"] = df["Email"].fillna("Not Provided")
df["Email"] = df["Email"].str.lower().str.strip()

# Review: fill missing, standardize text
df["Review"] = df["Review"].fillna("No Review")
df["Review"] = df["Review"].str.title().str.strip()

# Display cleaned text columns
df[["CustomerName", "Email", "Review"]]
```

### Step 6a: TotalPrice

```python
df["TotalPrice"] = df["Price"] * df["Quantity"] - df["Discount"]
df[["Price", "Quantity", "Discount", "TotalPrice"]]
```

### Step 6b: DeliveryTime

```python
df["DeliveryTime"] = (df["DeliveryDate"] - df["ReceivedDate"]).dt.days
df[["ReceivedDate", "DeliveryDate", "DeliveryTime"]]
```

### Step 6c: Review Score

```python
review_map = {
    "No Review": 0,
    "Poor": 1,
    "Bad": 2,
    "Average": 3,
    "Good": 4,
    "Excellent": 5
}

df["ReviewScore"] = df["Review"].map(review_map)
df[["Review", "ReviewScore"]]
```

### Step 7: Export Cleaned Dataset

```python
# Export final cleaned dataset
df.to_csv("Cleaned_Customer_Data.csv", index=False)
df.to_excel("Cleaned_Customer_Data.xlsx", index=False)
print(df)
```

### Optional- Round off

```python
# Convert specific columns to integer
cols_to_int = ["Age", "Price", "Quantity", "Discount", "TotalPrice", "DeliveryTime", "ReviewScore"]

for col in cols_to_int:
    df[col] = df[col].round(0).astype(int)

# Display DataFrame
df
```
